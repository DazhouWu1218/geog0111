```python
from geog0111.get_snow_data import get_snow_data
from geog0111.regularise import regularise

import matplotlib.pyplot as plt

snow,doy = get_snow_data(2019)
weight = (snow <= 100).astype(float)
snow[snow>100]=0.
snow = snow/100.
```

```python
plt.imshow(weight.mean(axis=0))
plt.colorbar()
```

```python
plt.imshow(snow.mean(axis=0))
plt.colorbar()
```

```python
reg_snow = regularise(snow,weight,1)
plt.imshow(reg_snow.mean(axis=0))
plt.colorbar()
plt.figure()
plt.plot(reg_snow[:,100,100])
```

```python
reg_snow = regularise(snow,weight,2.0)
plt.imshow(reg_snow.std(axis=0))
plt.colorbar()
plt.figure()
plt.plot(reg_snow[:,100,100])
```

```python
reg_snow = regularise(snow,weight,3.0)
plt.imshow(reg_snow.std(axis=0))
plt.colorbar()
plt.figure()
plt.plot(reg_snow[:,100,100])
```

```python
reg_snow = regularise(snow,weight,5.0)
plt.imshow(reg_snow.std(axis=0))
plt.colorbar()
plt.figure()
plt.plot(reg_snow[:,100,100])
```

```python
reg_snow = regularise(snow,weight,4.0)
plt.imshow(reg_snow.std(axis=0))
plt.colorbar()
plt.figure()
plt.plot(reg_snow[:,100,100])
```
